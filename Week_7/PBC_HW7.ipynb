#### 第二題 - 輸入檔案/統一編號的搜尋
* 大部分的商號只會有統一編號。但如果是連鎖商店的話則會有店鋪的統一編號，以及所屬總公司的統編。
* 每個欄位寬度為20個字元，靠左對齊
* 營業人名稱 統一編號(總機構統一編號) 設立日期 資本額 營業地址
    * 設立日期應從民國轉換為西元年月日，並以YYYY-MM-DD格式顯示。
    * 資本額應以整數的格式顯示，每三位加一個逗點。

```python
import csv
from datetime import date

fileName = input()
idx = input()

search_list = []
parse_data = []


with open(fileName, 'r', encoding = 'utf-8') as file:
    # 打開在file_path的檔案，編碼是cp950
    for l in csv.reader(file, quotechar='"', delimiter=',',
                         quoting=csv.QUOTE_ALL, skipinitialspace=True):
        parse_data.append(l)

for row in range(len(parse_data)):
    num = parse_data[row][1]
    num_parent = parse_data[row][2]
    if((num_parent.find(idx) != -1) or (num.find(idx) != -1)):
        sp2 = "　"
        name = parse_data[row][3].ljust(20,sp2) 
        if(len(num_parent) == 8): # 有母公司的號碼
            num += "(" + num_parent + ")"
        date_tran = int(parse_data[row][5]) + 19110000
        date_tran = str(date_tran)
        set_date = date(int(date_tran[0:4]), int(date_tran[4:6]), int(date_tran[6:8])).isoformat()
        fund = int(parse_data[row][4])
        address = parse_data[row][0]
        search_list.append([name, num, set_date, fund, address])

if(len(search_list) == 0):
    print("NO_DATA_FOUND")
else:
    # 營業人名稱[0] 統一編號[1](總機構統一編號) 設立日期[2] 資本額[3] 營業地址[4]
    st_search_list = sorted(search_list, key = lambda s: (s[2], -s[3], s[1]))[0:20]
    out_format = '{:<20s} {:<20s} {:<20s} {:<20,} {:<20s}'
    if(len(search_list) <= 20):
        for d in st_search_list:
            print(out_format.format(*d))
    else:
        for d in st_search_list[0:20]:
            print(out_format.format(*d))
```

#### 第四題 - 關注行業與資本額的關係

```python
#營業地址,統一編號,總機構統一編號,營業人名稱,資本額,設立日期,組織別名稱,使用統一發票,行業代號,名稱,行業代號1,名稱1,行業代號2,名稱2,行業代號3,名稱3


def median(lst):
    n = len(lst)
    s = sorted(lst)
    return int((sum(s[n//2-1:n//2+1])/2.0, s[n//2])[n % 2]) if n else None


# Read input
fileName = input()
parse_data = []
text_list = []

with open(fileName, 'r', encoding = 'utf-8') as file:
    # 打開在file_path的檔案，編碼是cp950
    for l in csv.reader(file, quotechar='"', delimiter=',',
                         quoting=csv.QUOTE_ALL, skipinitialspace=True):
        parse_data.append(l)

print(parse_data)
        
# Go through all industry in data
print_dict = {}

for data in parse_data[1:]:
    for i in range(8,16,2):
        if(data[i] in print_dict):
            print_dict[data[i]]['capitals'].append(int(data[4]))
        else:
            print_dict[data[i]] = {
                'name': data[9],
                'capitals': [int(data[4])]
            }

# Find median
for mid in print_dict:
    print_dict[mid]['median'] = median(print_dict[mid]['capitals'])

# Sort by median and only use the first ten and print
#行業代號 (行業名稱): $金額; N=數量
for number, info in sorted(print_dict.items(), key=lambda x: (-x[1]['median'], x[0]))[:20]:
    print(f'{number} ({info["name"]}): ${info["median"]}; N={len(info["capitals"])}')
```

```python
# Q4
fileName = input()                            # file name and dir
file = open(fileName, 'r', encoding='utf-8')  # open file

firstLine = True  # is first line or not
data = []         # data after cleaning
res = {}          # list for capital

# data preprocessing
for line in file.readlines():
    if firstLine:
        firstLine = False
        continue
    
    isText = False
    start = 1
    lineList = []
    for i in range(len(line)):
        if line[i] == '"' and isText == False:
            isText = True
        elif line[i] == '"' and isText == True:
            isText = False
        elif line[i] == ',' and isText == False:  # , isn't in the ""
            lineList.append(line[start:i])
            start = i+1
        elif i == len(line)-1:  # the last
            lineList.append(line[start:])
    data.append(lineList)
    
for d in data:
    for i in range(len(d)):
        d[i] = d[i].replace(" ", '').replace('\n', '').replace('"', '')

# create the pair dict
d = {}  # mapping id and name
res = {}  # store the pair
for t in data:
    temp = []  # store id
    for i in range(8, 15, 2):
        if t[i] != "":
            temp.append(t[i])
            if d.get(t[i]) == None:
                d[t[i]] = [t[i+1]]
    if len(temp) > 1:
        temp = sorted(temp) # sort from small to large
        for i in range(len(temp)-1):
            for j in range(i+1, len(temp)):
                if res.get((temp[i], temp[j])) == None:
                    res[(temp[i], temp[j])] = 1
                else:
                    res[(temp[i], temp[j])] += 1

# sort by times
final = [[] for i in range(len(res))]  # final results
idx = 0
res = sorted(res.items(), key=lambda item: item[1], reverse=True)
for key, value in res:
    final[idx].append(key[0])
    final[idx].append(key[1])
    final[idx].append(value)
    idx += 1

# sorted by id1
final = final[0:50]
for i in range(1, len(final)):
    while i != 0 and (final[i][2] == final[i-1][2] and int(final[i][0]) < int(final[i-1][0])):
        final[i-1], final[i] = final[i], final[i-1]
        i -= 1 
        
# sort by id2
for i in range(1, len(final)):
    if final[i][2] == final[i-1][2] and final[i][0] == final[i-1][0]:
        if int(final[i][1]) < int(final[i-1][1]):
            final[i-1], final[i] = final[i], final[i-1]

n = 0
for t in final:
    if n == 20:
        break
    n += 1
    
    print(f"{d[t[0]][0]} {d[t[1]][0]} {t[2]}")
```

```python
text = str(input())
punctuation = "#$%&'*+,-./:;<=>?@[]^_`{¦}~" + '"'

for char in text:
    for punc in punctuation:
        if char == punc:
            text = text.replace(char, ' ')
text = text.split(' ')

for word in text:
    cnt=0
    acronym = []
    if '(' and ')' in word:
        i = text.index(word)  # 縮寫的index
        acronym = word[1:(len(word) - 1)]  # 縮寫
        len_acronym = len(acronym)  # 縮寫長度
        total_name = text[i - len(acronym): i]  # 取出縮寫前面的單字
        out_total_name = ''  # 要印出的
        if len(total_name) == 0:  # 如果縮寫前面並無單字則跳過 (AI) atr iaa
            continue
        for j in range(len(total_name) - 1):  # 確認單字首字大寫（到倒數第二個）使否等於縮寫
            if len(total_name[j]) != 0:
                if ((total_name[j]).capitalize())[0] == acronym[j]:
                    out_total_name += total_name[j] + ' '  # 若一樣則加入印處的全名字串
                    cnt+=1
                    continue
                else:
                    break
        if cnt == (len_acronym-1) and ((total_name[-1]).capitalize()): # length of string is cal by word not space
            out_total_name += total_name[-1]
            print(acronym + ': ' + out_total_name)
```

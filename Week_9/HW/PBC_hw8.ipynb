```python
import math
class Card:
  def __init__(self, suit, rank): 
    self.suit = suit
    self.rank = rank
  
  def ruleA(self):
    """有A"""
    score=0
    for a in self.rank:
      if(a==1):
        score+=5
    return score
  
  def ruleB(self):
    """兩兩成對"""
    score =0
    sort_r = sorted(self.rank)
    N=len(sort_r)
    tmp=sort_r[0]
    tmp_cnt=1
    n=2
    for i in range(1,N):
      if(tmp==sort_r[i]):
        tmp_cnt+=1
      else:
        if(tmp_cnt != 1):
          p = math.factorial(tmp_cnt)//(math.factorial(n)*math.factorial(tmp_cnt-n))
          score+=10*p
        tmp_cnt=1
        tmp=sort_r[i]
    
    if(tmp_cnt!=1):
      p = math.factorial(tmp_cnt)//(math.factorial(n)*math.factorial(tmp_cnt-n))
      score+=10*p

    return score

  def ruleC(self):
    """同花"""
    isSame=True
    tmp = self.suit[0]
    n = len(self.suit)
    for i in range(1,n):
      if(tmp != self.suit[i]):
        isSame = False
        break
    if(isSame):
      return 30
    else:
      return 0

  def ruleD(self):
    """順子"""
    sort_r = sorted(self.rank)
    isTrue=False
    n = len(sort_r)
    a = [1,2,11,12,13]
    b = [1,2,3,12,13]
    c = [1,10,11,12,13]
    d = [1,2,3,4,13]
    if(sort_r == a or sort_r ==b or sort_r == c or sort_r == d):
      isTrue = True
    else:
      tmp = sort_r[0]
      for i in range(1,n):
        if(tmp==sort_r[i]-1):
          tmp = sort_r[i]
          isTrue=True
        else:
          isTrue=False
          break
     
    if isTrue:
      return 50
    else:
      return 0

  def ruleE(self):
    """葫蘆"""
    sort_r = sorted(self.rank)
    n = len(sort_r)
    isTriple = False
    isDouble = False
    tmp=sort_r[0]
    tmp_cnt =1

    for i in range(1,n):
      if(tmp == sort_r[i]):
        tmp_cnt+=1
      else:
        if(tmp_cnt==2):
          isDouble=True
        elif(tmp_cnt==3):
          isTriple = True
        tmp_cnt=1
      tmp = sort_r[i]  
    
    if(tmp_cnt==2):
      isDouble=True
    elif(tmp_cnt==3):
      isTriple = True

    if isTriple and isDouble:
      return 80
    else:
      return 0

  def ruleF(self):
    """四條"""
    sort_r = sorted(self.rank)
    n = len(sort_r)
    isQuadra = False
    tmp=sort_r[0]
    tmp_cnt =1

    for i in range(1,n):
      if(tmp == sort_r[i]):
        tmp_cnt+=1
      else:
        if(tmp_cnt==4):
          isQuadra=True
        tmp_cnt=1
      tmp = sort_r[i]  
    
    if(tmp_cnt==4):
      isQuadra=True

    if isQuadra:
      return 100
    elif(tmp_cnt==5):
      return 500
    else:
      return 0
  
  def ruleG(self):
    """同花順"""


    ## 順子
    sort_r = sorted(self.rank)
    isTrue=False
    n = len(sort_r)
    a = [1,2,11,12,13]
    b = [1,2,3,12,13]
    c = [1,10,11,12,13]
    d = [1,2,3,4,13]   
    if(sort_r == a or sort_r ==b or sort_r == c or sort_r == d):
      isTrue = True
    else:
      tmp = sort_r[0]
      for i in range(1,n):
        if(tmp==sort_r[i]-1):
          tmp = sort_r[i]
          isTrue=True
        else:
          isTrue=False
          break
    
    ## 同花
    isSame=True
    tmp = self.suit[0]
    for i in range(1,n):
      if(tmp==self.suit[i]):
        tmp = self.suit[i]
      else:
        isSame=False
        break
    
    if isTrue and isSame:
      return 300
    else:
      return 0

#### input
s_l = input().split(',')
tmp_r_l = input().split(',')
r_l = []
for c in tmp_r_l:
    if c == 'A':
        r_l.append(1)
    elif c == 'J':
        r_l.append(11)
    elif c == 'Q':
        r_l.append(12)
    elif c == 'K':
        r_l.append(13)
    else:
        r_l.append(int(c))

total_score =0
total_score = Card(s_l,r_l).ruleA()+Card(s_l,r_l).ruleB()+Card(s_l,r_l).ruleC()+Card(s_l,r_l).ruleD()+Card(s_l,r_l).ruleE()+Card(s_l,r_l).ruleF()+Card(s_l,r_l).ruleG()
print(total_score)
```
